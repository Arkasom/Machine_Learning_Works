# Splitting the Dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("stock_prices_dataset.csv")
print(df)
```

```python
target_column = "close"
Y = df[target_column]
X = df.drop(target_column,axis=1)
```

```python
X
```

```python
numerical_cols = [col for col in X.select_dtypes(include=[np.number])]
print(numerical_cols)
```

```python
categorical_cols = [col for col in X.select_dtypes(include='object')]
print(categorical_cols)
```

```python
X['date'] = pd.to_datetime(X['date'],errors='coerce')
print(X)
```

```python
X['year'] = X['date'].dt.year
X['month'] = X['date'].dt.month
X['day'] = X['date'].dt.day
X = X.drop('date', axis=1)
```

```python
X.dtypes
```

# handling Missing values

## using the 'fillna' method with median value for numerical columns

```python
for col in numerical_cols:
  if X[col].isnull().sum() > 0:
    median_value = X[col].median()
    X[col] = X[col].fillna(median_value)
    print(f"{col} filled with {median_value}")
```

## using the 'simpleimputer' with most-frequent strategy which is same as mode value

```python
imputer = SimpleImputer(strategy="most_frequent", missing_values=np.nan)
categorical_cols = [col for col in X.select_dtypes(include='object')]
for col in categorical_cols:
    if X[col].isnull().sum() > 0 or (X[col] == '').sum() > 0:
        X[[col]] = imputer.fit_transform(X[[col]])
        mode_value = imputer.statistics_[0]
        print(f"{col} is filled with {mode_value}")
```

```python
# X[numerical_cols].hist()
```

```python
# Check the number of unique values in the 'symbol' column
print(X['symbol'].nunique())
```

# One hot encoding

```python
X = pd.get_dummies(X, columns=['symbol'], drop_first=True)
display(X.head())
```

```python
X.shape
```

# Normalize numerical features
# We will use StandardScaler to standardize features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

display(X.head())
```

# Split the dataset into training and testing sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

# Train a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, Y_train)

print("Linear Regression model trained successfully!")
```

# Evaluate the Model

```python
# Make predictions on the test data
Y_pred = model.predict(X_test)

# Evaluate the model
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)

print(f"R-squared: {r2}")
print(f"Mean Squared Error (MSE): {mse}")
```

# Interpret the Model Coefficients

```python
# Get the coefficients and feature names
coefficients = model.coef_
feature_names = X.columns

# Create a pandas Series for easier viewing
coefficients_series = pd.Series(coefficients, index=feature_names)

# Print the coefficients
print("Model Coefficients:")
print(coefficients_series)

# Print the intercept
print("\nIntercept:", model.intercept_)
```
